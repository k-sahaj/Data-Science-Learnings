```python
def initialize_parameters(layer_dims):

  np.random.seed(3)
  parameters = {}
  L = len(layer_dims)

  for l in range(1, L):

     parameters['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
     parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))

  return parameters
```

```python
initialize_parameters([2,2,1])
```

we defined a function to initialize parameters. there are many approaches like giving some random value, giving 1 to W and 0 to b, scaling it down, etc.

In code,np.random.seed ensures ki har baar same hi random number generate ho i.e. If someone else runs the same code with the same NumPy version and the same seed, they'll get the same sequence.

shuru me parameters kuchh naii aur length is length.

loop me har l ke liye weights nikalne hain. Kon se l me kitna hoga decided from that (layer[--], layer[--])

```python
def linear_forward(A_prev, W, b):

  Z = np.dot(W.T, A_prev) + b

  return Z
```

Iss function me we are just calculating the Oxx i.e. z = (weight) ka transpose multiplied by Previous ka Output + bias


```python
#Forward Propagation

def L_layer_forward(X, parameters):

  A = X
  L = len(parameters) // 2 # idhar // ensures integer value

  for l in range(1, L+1):
    A_prev = A
    W1 = parameters['W' + str(l)]
    b1 = parameters['b' + str(l)]

    # print("A"+str(l-1)+": ", A_prev)
    # print("W"+str(l)+": ", W1)
    # print("b"+str(l)+": ", b1)
    # print("--"*20)

    A = linear_forward(A_prev, W1, b1)
    # print("A"+str(l)+": ", A)
    # print("**"*20)

  return A, A_prev
```

Idhar A's are nothing but collection of all outputs in that layer. A_prev ke liye pichhle layer ka, and we'll find out iss layer ka A using linear_forward wala function that we just created one cell above.

Thus, this function will return two things-
1. A which is the output
2. A previous which is the output of previous layer and we need that for our partial derivative calculation later in GD and updating.

```python
#Example check:
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1)
y = df[['lpa']].values[0][0]

parameters = initialize_parameters([2,2,1])
y_hat,A1 = L_layer_forward(X, parameters)
```

```python
y_hat = y_hat[0][0]
```

```python
def update_parameters(parameters, y, y_hat, A1, X):
  #Iss example me apne paas 9 parameters honge
  #lr = 0.001

  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * 2 * (y - y_hat)*A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat)*A1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * 2 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0]*X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][0][0])

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0]*X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * 2 * (y - y_hat)*parameters['W2'][1][0])

  #formula me - - = + ho gayaa
  return parameters
```

```python
update_parameters(parameters, y, y_hat, A1, X)
```

Using this W and b's that we got, do the same thig with other 3 row data changing the w and b's every single time little by little, also calculating loss for every row.

Then do the same thing for a no. of epochs (outer loop) and calculating loss of each epochs. Mind that har epoch ke saath loss decrease hona chaiye

```python
#Sidha epoch implementaion from start

parameters = initialize_parameters([2,2,1])
epochs = 5

for i in range(epochs):

  loss = []
  for j in range(df.shape[0]):

    X = df[['cgpa', 'profile_score']].values[j].reshape(2,1)
    y = df[['lpa']].values[j][0]

    #Parameter initialization

    y_hat, A1 = L_layer_forward(X, parameters)
    y_hat = y_hat[0][0]

    update_parameters(parameters, y, y_hat, A1, X)

    loss.append(np.square(y-y_hat))

  print('Epoch - ', i+1, 'Loss = ', np.array(loss).mean())
```

# **Backpropagation for Classification problems**

Change-
1. activation function taken sigmoid instead of linear

2. loss function taken binary cross entropy instead of mse

```python
def initialize_parameters(layer_dims):

  np.random.seed(3)
  parameters = {}
  L = len(layer_dims)

  for l in range(1, L):

     parameters['W' + str(l)] = np.ones((layer_dims[l-1], layer_dims[l]))*0.1
     parameters['b' + str(l)] = np.zeros((layer_dims[l], 1))

  return parameters
```

```python
#Utility function
def sigmoid(Z):

  A = 1/(1+np.exp(-Z))

  return A
```

```python
def sigmoid_forward(A_prev, W, b):

  Z = np.dot(W.T, A_prev) + b
  A = sigmoid(Z)

  return A
```

```python
def S_layer_forward(X, parameters):

  A = X
  L = len(parameters) // 2  #no. of layers in the neural network

  for l in range(1, L+1):
    A_prev = A
    W1 = parameters['W' + str(l)]
    b1 = parameters['b' + str(l)]

    A = sigmoid_forward(A_prev, W1, b1)

  return A, A_prev
```

```python
def update_parameters(parameters, y, y_hat, A1, X):
  # notes dekho uske formulaes hi code me badle hain

  parameters['W2'][0][0] = parameters['W2'][0][0] + (0.001 * (y - y_hat)*A1[0][0])
  parameters['W2'][1][0] = parameters['W2'][1][0] + (0.001 * (y - y_hat)*A1[1][0])
  parameters['b2'][0][0] = parameters['W2'][1][0] + (0.001 * (y - y_hat))

  parameters['W1'][0][0] = parameters['W1'][0][0] + (0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[0][0] * (1-A1[0][0]) * X[0][0])
  parameters['W1'][0][1] = parameters['W1'][0][1] + (0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[0][0] * (1-A1[0][0]) * X[1][0])
  parameters['b1'][0][0] = parameters['b1'][0][0] + (0.001 * (y - y_hat) * parameters['W2'][0][0] * A1[0][0] * (1-A1[0][0]))

  parameters['W1'][1][0] = parameters['W1'][1][0] + (0.001 * (y - y_hat) * parameters['W2'][1][0] *A1[1][0] * (1-A1[1][0]) * X[0][0])
  parameters['W1'][1][1] = parameters['W1'][1][1] + (0.001 * (y - y_hat) * parameters['W2'][1][0] *A1[1][0] * (1-A1[1][0]) * X[1][0])
  parameters['b1'][1][0] = parameters['b1'][1][0] + (0.001 * (y - y_hat) * parameters['W2'][1][0] *A1[1][0] * (1-A1[1][0]))

  return parameters
```

```python
X = df[['cgpa', 'profile_score']].values[0].reshape(2,1)
y = df[['lpa']].values[0][0]

parameters = initialize_parameters([2,2,1])

y_hat,A1 = S_layer_forward(X, parameters)
y_hat = y_hat[0][0]

update_parameters(parameters, y, y_hat, A1, X)

print('Loss individual: ', -(y*np.log(y_hat) + (1-y)*np.log(1-y_hat)))

parameters
```
